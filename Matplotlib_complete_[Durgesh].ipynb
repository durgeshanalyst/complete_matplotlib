# Label and title

#### Facts about font propertires for labela nd title

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.ylabel('Calorie Burnage',fontdict=font2)
plt.xlabel('Average Pulse',font2)
plt.title('sports watch data',font1,size=12,loc='left')
plt.plot(x,y)
plt.grid(axis='y',c='g',linestyle='--')
plt.show()
#fontdict for font style
#loc= left,right for font location
#size for resize
```

### Subplot()

subplot take 3 argumnet subplot=(row,column,when to plot number)

use to draw multiple plots in same figure

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#each point color
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.scatter(x, y, c=colors,cmap='viridis')
plt.grid(False)
plt.colorbar()
plt.show()
```

Creating Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors=['#DC143C','y','r','#C21E56']
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title='Four Fruits')
plt.show() 

# we have mylabels for tagging,startangle for rotate and explode= darar, shadow for provide shadow,
```
